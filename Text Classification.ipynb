# Using Manual Functions

```python
from sklearn.datasets import fetch_20newsgroups
from sklearn.metrics import accuracy_score as ac
```

```python
import numpy as np
```

```python
# fetching training data
train_data = fetch_20newsgroups(subset='train')     
```

```python
#print(len(train_data.data))
```

```python
#print(len(train_data.target))
```

```python
# all the documents of training dataset are stored
data = train_data.data                   
```

```python
# indexes of different categories are stored
y_indexs = train_data.target             
```

```python
import re
```

```python
# making list which contains all the words in the whole text data we have (i.e.,in all the documents)
li=[]
for i in range(len(data)):          # extracting only words from each document (ignoring the special characters)
    res = re.findall(r'\w+', data[i])    # res is a list which contains all the words of a document
    li+= res                             # concatenating or joining the list of each document into a single list
```

```python
#len(li)
```

```python
# lists which contains the stopwords
stop_w2=[]
stop_w=["i", "me", "my", "myself", "we", "our", "ours", "ourselves", "you", "your", "yours", "yourself", "yourselves", "he", "him", "his", "himself", "she", "her", "hers", "herself", "it", "its", "itself", "they", "them", "their", "theirs", "themselves", "what", "which", "who", "whom", "this", "that", "these", "those", "am", "is", "are", "was", "were", "be", "been", "being", "have", "has", "had", "having", "do", "does", "did", "doing", "a", "an", "the", "and", "but", "if", "or", "because", "as", "until", "while", "of", "at", "by", "for", "with", "about", "against", "between", "into", "through", "during", "before", "after", "above", "below", "to", "from", "up", "down", "in", "out", "on", "off", "over", "under", "again", "further", "then", "once", "here", "there", "when", "where", "why", "how", "all", "any", "both", "each", "few", "more", "most", "other", "some", "such", "no", "nor", "not", "only", "own", "same", "so", "than", "too", "very", "s", "t", "can", "will", "just", "don", "should", "now"]
for i in stop_w:
    stop_w2.append(i.title())
    
# a list which contains words which are occuring in all the documents and some special characters
stop_w3=["From","Organization","Subject","Lines","Nntp-Posting-Host","___","__","____"]
```

```python
d={}
```

```python
# making a dictionary which doesn't contain the words present in above 3 lists(stop_w, stop_w2, stop_w3) 
# contains all the unique words along with their frequencies
for w in li:
    if w not in stop_w:
        if w not in stop_w2:
            if w not in stop_w3:
                if len(w)!=1:
                    d[w]=d.get(w,0)+1
```

```python
#d
```

```python
# di is a list which contains the key and its frequency (in a tuple) as elements, and they are sorted in reverse order by values of dictionary d
di=sorted(d.items(), key=lambda x: x[1], reverse=True)
```

```python
#di
```

```python
#len(di)
```

```python
# selecting the top 3000 words with maximum frequency as our feature set
features=[]
for k in range(3000):
    features.append(di[k][0])
```

```python
# returns a 2d list which contains the frequency of the features in each document
# calling this funtion for training data and as well as for testing data
def frequency_of_features_in_each_document(data):
    s=[]
    for i in range(len(data)):
        dic={}
        l = re.findall(r'\w+', data[i])     # a list of all the words present in a document
        # making a dictionary which contains the feature name and its frequency in ith document 
        for k in features:
            if k in l:
                dic[k]=dic.get(k,0)+1
            else:
                dic[k]=0
        # appending the list of frequencies of all feature of ith doc in the final 2-d list
        s.append(list(dic.values()))        # transformed the text data into numerical data
    return s
```

```python
# a 2d list which contains the frequency of the features in each train document
result= frequency_of_features_in_each_document(data)
```

```python
# df is the training dataset
import pandas as pd
df = pd.DataFrame(result, columns=features)
```

```python
data_test = test_data.data
y_indexs_test = test_data.target
```

```python
# a 2d list which contains the frequency of the features in each test document
result_test=frequency_of_features_in_each_document(data_test)
```

```python
# df_test is the testing dataset
df_test = pd.DataFrame(result_test, columns=features)
```

```python
df_test
```

```python
# fitting the training data 
# built a dictionary having 3 levels: 1.all the classes, 2.features in each class, 3.no. of words in each feature of each class
def fit(x,y):
    dictionary={}
    dictionary["total_data"]=x[:,:].sum()   # key containing the total no. of words in all the classes 
    for i in set(y):                        # iterating over all categories
        dictionary[i]={}
        indexs= y==i                        
        x_current= x[indexs]                # x_current contain only those documents which has i as its category
        dictionary[i]["total_count"]=x_current[:,:].sum()   # key containing the no. of words in ith category
        for j in range(3000):               # iterating over all features
            dictionary[i][j]=x_current[:,j].sum()      # for each feature counting the total no. of words (in ith category)
    return dictionary
```

```python
dict_fit= fit(df.values,y_indexs)
```

```python
# probabilily function which calculates the log probability of a test document(x) when the category is k
def prob(d,x,k):
    op= np.log(d[k]["total_count"])-np.log(d["total_data"])  # log probability of kth category
    # calculating the probablity of only those words which are present in the vocabulary 
    for j in range(3000):
        xj=x[j]
        if xj!=0:                         # if xj(i.e, the frequency of word) is equal to 0 then that means that particular word isn't present in the current document, so ignorong such words
            c=d[k][j]+1                   # total no. of words in feature j when category is k along with laplace correction
            c1=d[k]["total_count"]+len(features)
            cp=np.log(c)-np.log(c1)
            op=op+cp
    return op
```

```python
def predictsingle(d,x):                      
    classes=d.keys()
    best_p=-1
    best_class=-1
    first_run = True
    for k in classes:                # iterating over each category and passing them to the probability function 
        if k!="total_data":
            p_current=prob(d,x,k)    # log probability of the test document when the category is k
            # finding the maximum probability of a test document and the category with which its probability is max
            if first_run or p_current > best_p:
                best_class=k
                best_p= p_current
            first_run = False
    return best_class                # returning the category of the test document
```

```python
# function to predict the category of each test document and appending it to a list y_pred
def predict(d,x_test):
    y_pred=[]
    for x in x_test:          
        y_pred.append(predictsingle(d,x))     # passing each test document to predictsingle funtion one by one
    return y_pred
```

```python
y_p=predict(dict_fit,df_test.values)
```

```python
# confusion matrix
from sklearn.metrics import confusion_matrix as cm
print(cm(y_indexs_test,y_p))
```

```python
# accuracy
print("Accuracy=",ac(y_indexs_test,y_p))
```
